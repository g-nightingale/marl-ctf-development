# Analysis

```python
from gridworld_ctf import GridworldCtf
import numpy as np
from scenarios import CtfScenarios as scn
from utils import get_agents, create_gif, create_plots, plot_heatmaps, \
                  duel, TrainingConfig
```

### Experiment 1: Arrow

```python
# Create training plots
create_plots('runs/0_arrow/',
            ['time_capsule_0_50.bin',
             'time_capsule_1_50.bin',
             'time_capsule_2_50.bin',
             'time_capsule_3_50.bin',
             'time_capsule_4_50.bin',
             'time_capsule_5_50.bin',
             'time_capsule_6_50.bin',
             'time_capsule_7_50.bin',
             'time_capsule_8_50.bin',
             'time_capsule_9_50.bin'])
```

```python
# Retrieve agents from time capsule
env_config, agent1, agent2 = get_agents('runs/0_arrow/',
                                        'time_capsule_0_50.bin',
                                        0,
                                        0,
                                        1,
                                        0)

# Create environment with loaded config                         
env = GridworldCtf(**env_config)

# Render a duel
_, _, _ = duel(env, 
                agent1, 
                agent2, 
                (0, 1), 
                max_steps=500, 
                render=True,
                sleep_time=0.05)
```

```python
# Plot heatmaps from the latest run
plot_heatmaps(env)
```

```python
# Generate an animated gif
create_gif(env, 
            agent1, 
            agent2, 
            max_frames=250,
            gif_folder='/gifs/0_arrow/',
            gif_name='arrow.gif',
            delete_temp_files=False,
            duration=200)
```
